```python
import cv2
import pytesseract
import logging
import session_info
from pdf2image import convert_from_path

session_info.show()
```

# For JPEG, PNG

```python
# create loger
logging.basicConfig(level=logging.INFO)
loger = logging.getLogger('tesseract_info')

# path to tesseract
pytesseract.pytesseract.tesseract_cmd = '.../Tesseract-OCR/tesseract.exe'
tessdata_dir_config = '.../Tesseract-OCR/tessdata'

input_dir = '.../test_scan.jpg'

# photo
img = cv2.imread(input_dir)
loger.info('reed photo')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
loger.info('photo to RGB')

# Creating a text file to write the output
outfile = ".../out_jpg_text.txt"
loger.info('create .txt file')
  
# Open the file in append mode so that 
# All contents of all images are added to the same file
f = open(outfile, "a", encoding="utf-8")
loger.info('file open')
          
# Recognize the text as string in image using pytesserct
text = pytesseract.image_to_string(img, config=tessdata_dir_config)
loger.info('text from image -> to string')

# write the processed text to the file.
f.write(text)
loger.info('write text from image to file')

# Close the file after writing all the text.
f.close()
loger.info('close file')
```

## For PDF files 

```python
# Path of the pdf
pdf_file = ".../200300744v1_image_pdf.pdf"
tessdata_dir_config = '.../Tesseract-OCR/tessdata'
  
'''
Converting PDF to images
'''
  
# Store all the pages of the PDF in a variable
pages = convert_from_path(pdf_file, 500)
loger.info('convert pages from pdf file')
  
# Counter to store images of each page of PDF to image
image_counter = 1
  
# Iterate through all the pages stored above
for page in pages:
  
    # Declaring filename for each page of PDF as JPG
    # PDF page 1 -> page_1.jpg
    # PDF page n -> page_n.jpg
    filename = "page_"+str(image_counter)+".jpg"
      
    # Save the image of the page in system
    page.save(filename, 'JPEG')
    loger.info('save .jpg photo')
  
    # Increment the counter to update filename
    image_counter = image_counter + 1

'''
images to text
'''

# Variable to get count of total number of pages
filelimit = image_counter-1
  
# Creating a text file to write the output
outfile = ".../out_text.txt"
loger.info('create output file')
  
# Open the file in append mode so that 
# All contents of all images are added to the same file
f = open(outfile, "a", encoding="utf-8")
loger.info('file open')
  
# Iterate from 1 to total number of pages
for i in range(1, filelimit + 1):
  
    # Set filename to recognize text from
    # page_1.jpg
    # page_n.jpg
    filename = "page_"+str(i)+".jpg"
          
    # Recognize the text as string in image using pytesserct
    text = pytesseract.image_to_string(filename, config=tessdata_dir_config)
    loger.info('text from image -> to string')

    # write the processed text to the file.
    f.write(text)
    loger.info('write text from image to file')
    
# Close the file after writing all the text.
f.close()
loger.info('close file')
```
